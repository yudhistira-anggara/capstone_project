# data cleaning and regroup of labels 
- in dermnet dataset in kaggle, each label is too general in case that each labels contain any diseases that have different characteristic. the purpose of this NoteBook is to parse a label into a spesific labels

## make new directory based on sub labels

```python
labels=[acnitic_keratosis, seborheic_keratosis, acne, systemic_disease, eczema, warts, ringworm, nail,Light_Diseases]
sub_labels=[]
for label in labels:
    sub_labels.extend(label)
```

```python
new_dataset=os.path.join('D:\\JupyterNotebook',"new_dataset")# pass directory here
new_train_dir=os.path.join(new_dataset,"train")
new_test_dir= os.path.join(new_dataset,"test")
```

```python
for i in sub_labels:
    # train
    os.makedirs(os.path.join(new_train_dir,i))
    
    #test
    os.makedirs(os.path.join(new_test_dir,i))
    print(i)
```
